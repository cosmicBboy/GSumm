```python
from gensim import corpora, models, similarities

corpus = [[(0, 1.0), (1, 1.0), (2, 1.0)],
          [(2, 1.0), (3, 1.0), (4, 1.0), (5, 1.0), (6, 1.0), (8, 1.0)],
          [(1, 1.0), (3, 1.0), (4, 1.0), (7, 1.0)],
          [(0, 1.0), (4, 2.0), (7, 1.0)],
          [(3, 1.0), (5, 1.0), (6, 1.0)],
          [(9, 1.0)],
          [(9, 1.0), (10, 1.0)],
          [(9, 1.0), (10, 1.0), (11, 1.0)],
          [(8, 1.0), (10, 1.0), (11, 1.0)]]
```

```python
tfidf = models.TfidfModel(corpus)
```

```python
vec = [(0,1), (4,1)]
print(tfidf[vec])
```

```python
index = similarities.SparseMatrixSimilarity(tfidf[corpus], num_features=12)
```

```python
#computes similarity between vec and document 0-9
sims = index[tfidf[vec]]
print(list(enumerate(sims)))
```

```python
#define a class MyCorpus that iterates through corpus.txt file
#for more memory efficient way of grabbing documents from corpus
class MyCorpus(object):
    def __init__(self, filePath):
        self.filePath = filePath
    
    def __iter__(self):
        for line in open(self.filePath, 'r'):
            yield dictionary.doc2bow(line.lower().split())
```

```python
#to construct the dictionary without loading all text into memory
from nltk.corpus import stopwords

dictionary = corpora.Dictionary(line.lower().split() for line in open('../smallData/seedCorpus.txt'))
stop_ids = [dictionary.token2id[stopword] for stopword in stopwords.words()
            if stopword in dictionary.token2id]
once_ids = [tokenid for tokenid, docfreq in dictionary.dfs.iteritems() if docfreq == 1]
dictionary.filter_tokens(stop_ids + once_ids)
dictionary.compactify()
print(dictionary)
```

```python
corpusIter = MyCorpus('../smallData/seedCorpus.txt')
tfidf = models.TfidfModel([doc for doc in corpusIter])
```

```python
index = similarities.SparseMatrixSimilarity(tfidf[[doc for doc in corpusIter]], num_features= 3855)
```

```python
sims = index[tfidf[corpus[24]]]
print(list(enumerate(sims)))
```
